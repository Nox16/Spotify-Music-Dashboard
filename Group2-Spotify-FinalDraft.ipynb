### Playlist Inclusions per Decade

#### Guiding Question
I wanted to figure out what decade of music had the most playlist inclusions, as people often organize their music by decade.

To do, I opted to create a boxplot of these values to compare the total amount, average, median, and quartiles in one plot.

#### Finding Each Song's Release Decade

```python
# Create Numpy List of Decades from 1960 - 2020.
years = np.arange(1960, 2023, step=10)

# Copy New DataFrame with only the Release Year & Number of Spotify Playlist Inclusions
yearsPlaylist = music[["released_year", "in_spotify_playlists"]].copy()

# Make a new feature 'Decade' which is the release year minus the single's digit, which simplifies
# each song's release year to just it's decade.
yearsPlaylist["decade"] = yearsPlaylist["released_year"] - (yearsPlaylist["released_year"] % 10)

# Drop the Release Year column since it will no longer be used.
yearsPlaylist = yearsPlaylist.drop(columns=["released_year"])

# Remove any entries with a Decade lower than 1960 (due to a lack of worthwhile data)
yearsPlaylist = yearsPlaylist[yearsPlaylist["decade"] >= 1960]
```

#### Plotting a Boxchart of Decades with Playlist Inclusions

```python
# Set Size of Figure
fig, ax = plt.subplots(figsize=(10, 8))

# Set Grid and Legend Beneath Boxplot
ax.set_axisbelow(True)

"""
Manual Legend Source:
https://stackoverflow.com/questions/45786714/custom-marker-edge-style-in-manual-legend
"""

# Import 'mlines' to Create Manual Legend
import matplotlib.lines as mlines

# Set Legend Elements to White Dot with Label
white_dot = mlines.Line2D([], [], color='white', markeredgecolor = "black", marker='o',
                          markersize=10, label='Average Inclusions per Decade')

# Show Manual Legend
ax.legend(handles=[white_dot])

# Create Boxplot of Decade v. Playlist Inclusions
sns.boxplot(
    # X Values are the Decades, Y Values are the Playlisy Inclusions
    x=yearsPlaylist["decade"], y=yearsPlaylist["in_spotify_playlists"],

    # Show Markers for Mean Value & Remove Outliers
    showmeans=True, showfliers = False,

    # Format Marker as White Dot with Black Outline
    meanprops={
            "marker": "o",
            "markeredgecolor": "black",
            "markerfacecolor" : "white",
            "markersize": "10",
            })

# Figure Labeling
plt.title("Number of Playlist Inclusions per Decade")
plt.xlabel("Decade")
plt.ylabel("Number of Playlist Inclusions")

# Set Figure Background Grid
plt.grid(color='black', linestyle='-', linewidth=.25, zorder=0)

# Show Figure
plt.show()
```

#### Conclusion: Playlist Inclusions per Decade
From 1960 to 2020, I've found that:
- The 1990's have the highest average and highest extreme of inclusions.
- The 2000's have the highest median inclusions.
- The 1990's have the lowest median.
- The 2020's have the lowest average, most likely due to a lack of songs. The 1960's have the second lowest average.

### Playlist Inclusions per Song Attributes

#### Guiding Question
I wanted to find how four specific song attributes affected the number of playlists a song is added to.

These attributes are:
- Danceability (%)
- Energy (%)
- Liveness (%)
- Speechiness (%)

Each contribute something to how a song is received and may affect how often a song is added to playlists.

The best way to see the relationship between the number of playlist inclusions and these attributes would be a scatterplot per inclusion.

```python
# Make List of Attributes to Plot
attributes = ["danceability_%", "energy_%", "liveness_%", "speechiness_%"]

# Make DataFrame of Playlist Inclusions, Stream Count, & Attributes
attributePlaylists = music[["in_spotify_playlists", "streams", "danceability_%", "energy_%", "liveness_%", "speechiness_%"]]

# Create Subplots of 2x2 Figures, with an overall size of 16x12
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Flattens Axes Ararys
axes = axes.flatten()

# Y Values are the Number of Spotify Playlist Inclusions without 0's
y = attributePlaylists['in_spotify_playlists'].replace(0, np.nan)

# Loop through Attributes
for i, column in enumerate(attributes):
    # X Values will be the Feature of the given Attribute
    x = attributePlaylists[column]

    # Set X Values to be Multiples of 5, for cleaner plotting
    x = x.apply(lambda x: x - (x % 5))

    # Set X-Ticks to be from 0 to 100 with steps of 5 (0, 5, 10, ...)
    plt.xticks(np.arange(0, 100, step = 5))

    # Set Current Axes to the Axes[Index]
    plt.sca(axes[i])

    # Plot Scatter per Attribute, with a colormap based on the number of streams per song
    plt.scatter(x, y, c=attributePlaylists["streams"], cmap="viridis", s=50)

    # Set Colorbar of Streams with Label
    plt.colorbar(label="Streams")

    # Find Linear Regression Values per Attribute Plot
    slope, intercept, r, p, se = linregress(x, y)
    calc = (slope * x) + intercept

    # Plot Linear Regression (Line of Best Fit) per Attribute
    plt.plot(x, calc, color="red", alpha=1)

    # Edit Label to Remove "_%" Characters
    newLabel = column.capitalize().replace("_%", "")

    # Figure Labeling & Titling
    plt.xlabel(newLabel, fontsize=12)
    plt.ylabel('Spotify Playlists', fontsize=12)
    plt.title(f'Spotify Playlists vs. {newLabel}', fontsize=14)

# Figure Layout Editing & Showing
plt.tight_layout()
plt.show()
```

#### Conclusion: Playlist Inclusions per Song Attributes
Each scatter shows that none of these four attributes actually have that much of an impact on the number of playlist inclusions.

Danceability, Liveness, and Speechiness all suggest the higher the attribute the less playlist inclusions will occur, which goes against most assumptions. However, that could be because not many songs are rated that high with those attributes.

Energy has a positive relationship, but only marignally.

### Playlist Inclusion Percentage by Platform (Spotify v. Apple)

#### Guiding Question
I wanted to find what platform between Spotify and Apple Music had more playlist inclusions for the top 100 streamed songs.

Since the dataset is centered around Spotify, taking the raw values would unfairly favor Spotify. To get around this, I calculated the percent inclusion per song, by finding how many inclusions a specfic song had over the total playlist inclusions for a given platform.

```python
# Sort Songs by Stream Count
sortedStreams = music.sort_values(by=["streams"], ascending=False)

# Platform Playlists
spotPlay = music["in_spotify_playlists"]
appPlay = music["in_apple_playlists"]

# Total Playlist Inclusions
totalSpotPlay = spotPlay.sum()
totalAppPlay = appPlay.sum()

# Find Percent Inclusion per Song per Platform
sortedStreams["Spotify Playlist Inclusion (%)"] = round(((spotPlay / totalSpotPlay) * 100), 3)
sortedStreams["Apple Playlist Inclusion (%)"] = round(((appPlay / totalAppPlay) * 100), 3)

# Get Top 100 Songs by Streams
topStreams = sortedStreams.copy().head(100)
```

```python
# Set Figure Size to 10x8
plt.figure(figsize=(10,8))

# On the Scatter Plot, Plot a line between each marker of the same song for each playform.
x = topStreams["streams"]
y_spotify = topStreams["Spotify Playlist Inclusion (%)"]
y_apple = topStreams["Apple Playlist Inclusion (%)"]
plt.plot([x, x], [y_spotify, y_apple],
         linestyle='dotted', linewidth=0.5,
         color='blue', alpha=0.2,
         zorder=0)

# Scatter Plot of Spotify's Playlist Inclusion
plt.scatter(topStreams["streams"], topStreams["Spotify Playlist Inclusion (%)"], color='green', label='Spotify',)
# Scatter Plot of Apple's Playlist Inclusion
plt.scatter(topStreams["streams"], topStreams["Apple Playlist Inclusion (%)"], color='red', label='Apple')

# Using Linear Reression, find Line of Best Fit for Spotify (Plotted Green)
slope, intercept, r, p, se = linregress(topStreams["streams"], topStreams["Spotify Playlist Inclusion (%)"])
calc = (slope * topStreams["streams"]) + intercept
plt.plot(topStreams["streams"], calc,
         color="green", alpha=0.5)

# Using Linear Reression, find Line of Best Fit for Apple Music (Plotted Red)
slope, intercept, r, p, se = linregress(topStreams["streams"], topStreams["Apple Playlist Inclusion (%)"])
calc = (slope * topStreams["streams"]) + intercept
plt.plot(topStreams["streams"], calc,
         color="red", alpha=0.5)

# Figure Labeling
plt.title("Top 100 Songs Playlist Inclusion (%)")
plt.xlabel('Streams (Billions)')
plt.ylabel('Playlist Inclusion (%)')
plt.legend()

# Show plot
plt.show()
```

#### Conclusion: Playlist Inclusion Percentage by Platform (Spotify v. Apple)
From this scatter we can that when a song has less streams overall, it is typically in more on playlists on Spotify than Apple music.

This makes sense, as Spotify as a music platform is more accessible to smaller artists and has a wider selection, so popular songs without large streaming numbers may appear more on playlists.

At around ~2.7 Billion Streams Apple Music begins to have more playlist inclusions.

## **Topic: BPM (K-Means Clustering)**

### Elbow Plot

```python
df = music

# standardizing the BPM column
df_bpm = music[['bpm']].copy()
scaler = StandardScaler()
df_bpm_scaled = scaler.fit_transform(df_bpm)

# finding the ideal number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(df_bpm_scaled)
    inertia.append(kmeans.inertia_)

# plotting the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

The elbow plot shows that the inertia, sum of squared distances to the nearest cluster center, decreases as the number of clusters increases. The ideal number of cluster is found at the "elbow" point, where the rate of decrease sharply shifts. I determined this to be 3 clusters.

### Cluster Histogram

```python
# applying K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, n_init=10)
df_bpm['cluster'] = kmeans.fit_predict(df_bpm_scaled)

# creating a histogram to visualize the clusters
plt.figure(figsize=(10, 6), facecolor='white')
sns.histplot(df_bpm, x='bpm', hue='cluster', element='step', palette='viridis', bins=30)
plt.title('Distribution of BPM Across Clusters')
plt.xlabel('BPM')
plt.ylabel('Count')
plt.show()
```

The histogram shows how songs are grouped into three distinct BPM ranges. Each cluster represents a range of BPMs, showing variability in song tempos. This visualization can help in understanding the typical BPM ranges and their frequencies within the dataset.

### Relationship of Valence & Danceability

#### Guiding Question
What is the nature of the correlation between the valence and danceability of songs? Does a higher valence percentage tend to correlate with higher danceability, indicating that more positive songs are also more danceable?

```python
cmusic = music.copy()

cmusic.dropna(subset=['released_year', 'valence_%'], inplace=True)

#Calculate the mean for each year to generalize each data point
yearly_valence = cmusic.groupby('released_year')['valence_%'].mean()

plt.figure(figsize=(10, 6))
plt.plot(yearly_valence.index, yearly_valence.values, marker='o', color='blue', linestyle='-')
plt.title('Average Valence Percentage by Release Year on Spotify')
plt.xlabel('Release Year')
plt.ylabel('Average Valence Percentage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
cmusic.dropna(subset=['released_year', 'danceability_%'], inplace=True)

#Calculate the mean for each year to generalize each data point
av_dance = cmusic.groupby('released_year')['danceability_%'].mean()

plt.figure(figsize=(10, 6))
plt.plot(av_dance.index, av_dance.values, marker='o', color='red', linestyle='-')
plt.title('Average Danceability by Release Year on Spotify')
plt.xlabel('Release Year')
plt.ylabel('Average Danceability Percentage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
cmusic.dropna(subset=['danceability_%', 'valence_%'], inplace=True)

# Create a scatter plot to visualize the relationship between danceability and valence
plt.figure(figsize=(8, 6))
plt.scatter(music['valence_%'], cmusic['danceability_%'], color='black', alpha=0.5)
plt.title('Relationship Between Danceability and Valence on Spotify')
plt.xlabel('Danceability (%)')
plt.ylabel('Valence (%)')
sns.regplot(x='valence_%', y='danceability_%', data=cmusic,scatter_kws={'color':'skyblue'})
correlation_coefficient = music['danceability_%'].corr(music['valence_%'])
print("The correlation coefficient is:", correlation_coefficient)

plt.grid(True)
plt.tight_layout()
plt.show()
```

Before discussing the similarities between danceability and valence, its important to consider how the averages for these values changed over time. Interestingly, the line graphs correlating each of these variables with release dates turned out to be practically identical.

In conclusion, the analysis reveals a noteworthy similarity between danceability and valence, particularly in how their averages evolved over time, and a consistent positive trend between danceability and valence, indicating a potential relationship worth exploring further in music research.
